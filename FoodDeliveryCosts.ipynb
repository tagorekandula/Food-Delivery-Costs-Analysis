```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
df=pd.read_csv(r"C:\Users\Tagor\Downloads\food delivery costs.csv")
df
```

```python
df.info()
```

```python
df["Order Date and Time"]=pd.to_datetime(df["Order Date and Time"])
df.info()
```

```python
df["Delivery Date and Time"]=pd.to_datetime(df["Delivery Date and Time"])
df.info()
```

```python
df.head()
```

```python
def seperationofdiscounts(value):
    var=str(value).split(" ")[0]
    return var
df["Discounts and Offers"]=df["Discounts and Offers"].apply(seperationofdiscounts)
df
```

```python
def conversionofpercentage(value):
    if '%' in value:
        var = value.replace('%',"")
        return float(var)
    else:
        return float(value)
df["Discounts and Offers"]=df["Discounts and Offers"].apply(conversionofpercentage)
df
```

```python
df.loc[(df["Discounts and Offers"]<=15),"Discounts and Offers"]=(df["Discounts and Offers"]/100)*df["Order Value"]
df.head()
```

```python
df["Discounts and Offers"]=df["Discounts and Offers"].fillna(0)
df.head()
```

```python
df["Costs"]=df["Delivery Fee"]+df["Discounts and Offers"]+df["Payment Processing Fee"]
df.head()
```

```python
df["Profit"]=df["Commission Fee"]-df["Costs"]
df
```

```python
df["Profit"].sum()
```

```python
cost_dist=df[["Delivery Fee","Payment Processing Fee","Discounts and Offers"]].sum()
cost_dist
```

```python
plt.pie(cost_dist,labels=cost_dist.index,autopct="%1.1f%%",textprops={'color':'b'})
plt.show()
```
